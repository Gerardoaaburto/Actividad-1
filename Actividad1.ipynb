```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
diamonds = pd.read_csv('diamonds.csv')
diamonds
```

```python
#Identificaremos nulos y outliers para eliminarlos. 
```

```python
diamondsc=diamonds.copy()
```

```python
valores_nulos=diamondsc.isnull().sum()
valores_nulos
```

```python
diamondsc["carat"]=diamondsc["carat"].fillna(round(diamondsc["carat"].mean(),1))
diamondsc
```

```python
diamondsc["depth"]=diamondsc["depth"].fillna(round(diamondsc["depth"].mean(),1))
diamondsc
```

```python
diamondsc["table"]=diamondsc["table"].fillna(round(diamondsc["table"].mean(),1))
diamondsc
```

```python
diamondsc["price"]=diamondsc["price"].fillna(round(diamondsc["price"].median(),1))
diamondsc
```

```python
diamondsc["x"] =diamondsc["x"].fillna(method="ffill") 
diamondsc
```

```python
diamondsc["y"] =diamondsc["y"].fillna(method="ffill") 
diamondsc
```

```python
diamondsc["z"] =diamondsc["z"].fillna(method="ffill") 
diamondsc
```

```python
valores_nulos=diamondsc.isnull().sum()
valores_nulos
```

```python
#Outliers
```

```python
#Método aplicando desviación estandar. Encuentro los valores extremos
y=diamondsc
Limite_Superior= y.mean() + 3*y.std()
Limite_Inferior= y.mean() - 3*y.std()
print("Limite superior permitido", Limite_Superior)
print("Limite inferior permitido", Limite_Inferior)
```

```python
outliers= diamondsc[(y>Limite_Superior)|(y<Limite_Inferior)]
outliers
```

```python
#Obtenemos datos sin outliers del DataFrame
dt= diamondsc[(y<=Limite_Superior)&(y>=Limite_Inferior)]
dt
```

```python
#Eliminamos columnas innecesarias
dt= dt.drop(['color'], axis=1)
dt
```

```python
dt= dt.drop(['cut'], axis=1)
dt
```

```python
dt= dt.drop(['clarity'], axis=1)
dt
```

```python
#Reemplazamos valores atípicos (nulos) del dataframe con "mean"
#Realizamos una copia del dataframe
data_clean=dt.copy()
data_clean=data_clean.fillna(round(dt.mean(),1))
data_clean
```

```python
valores_nulos=data_clean.isnull().sum()
valores_nulos
```

```python
data=data_clean.copy()
```
